## Телефоны

```python
# заменяем Nan на []

cols = ['phone_admission', 'phone_general', 'email_general', 'email_admission']

for col in cols:
    df[col] = df[col].fillna('[]')

```

## Гос/не гос

```python
df.is_goverment.unique()
```

```python
mapping = {'Гос.вуз': 1, 'Коммерческий': 0}
df['is_goverment'] = df['is_goverment'].map(mapping)
```

## Рейтинг

```python
df['rating'].unique()
```

```python
rating_map = {
    "A+": 10,
    "A": 9,
    "A-": 8,
    "B+": 7,
    "B": 6,
    "B-": 5,
    "C+": 4,
    "C": 3,
    "C-": 2,
    "no data": -1
}

df['rating'] = df['rating'].map(rating_map)
df
```

```python
new = sorted(list(df['geolocation'].unique()))
new
```

```python
import ast

def count_items(cell):
    try:
        lst = ast.literal_eval(cell)
        if isinstance(lst, list):
            return len(lst)
        return 0
    except Exception:
        return 0

mask = (
    (df['phone_admission'].apply(count_items) > 3) |
    (df['phone_general'].apply(count_items) > 3) |
    (df['email_general'].apply(count_items) > 3) |
    (df['email_admission'].apply(count_items) > 3)
)

result = df[mask]
result
```
